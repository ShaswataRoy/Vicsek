```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('default')
#Extract Data
x1,y1 = np.loadtxt('vicsek40.txt',delimiter='\t',unpack=True)

x2,y2 = np.loadtxt('vicsek100.txt',delimiter='\t',unpack=True)

x3,y3 = np.loadtxt('vicsek400.txt',delimiter='\t',unpack=True)

x4,y4 = np.loadtxt('vicsek1000.txt',delimiter='\t',unpack=True)
x = np.hstack((x1,x2,x3))
y=np.hstack((y1,y2,y3))
plt.plot(x1,y1,'r')
plt.plot(x2,y2,'g')
plt.plot(x3,y3,'b')
plt.plot(x4,y4,'k')
```

```python
err = [abs(y1[i]-y1[12]) for i in [13,14,15,16]]
eta_change = [abs((x1[i]-x1[12])/x[12]) for i in [13,14,15,16]]
```
