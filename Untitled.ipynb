```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('default')
```

```python
for i in np.arange(199501,200000):
    L=128
    x,y,theta = np.loadtxt('data/'+str(i)+'.txt',delimiter='\t',unpack=True)

    axes = plt.gca()
    axes.set_xlim([0,L])
    axes.set_ylim([0,L])
    plt.quiver(x,y,np.cos(theta),np.sin(theta), units='width',scale=100,width=0.005)    #plt.scatter(x, y)
    plt.savefig('image/'+str(i)+'.png')
    plt.gcf().clear()
```

```python
n = len(x)
r = []
for i in range(n):
    for j in range(n):
        if(i!=j):
            r.append(np.sqrt((x[i]-x[j])**2+(y[i]-y[j])**2))
```

```python
import seaborn as sns
sns.distplot(r)
```

```python
r_bins = np.zeros(int(np.max(r))+1)
for i in range(len(r)):
    r_bins[int(r[i])]+=1
for i in range(int(np.max(r))+1):
    r_bins[i] = r_bins[i]/(2*np.pi*(i+0.5))
plt.plot(r_bins)
```

```python
r=[]
for i in np.arange(199900,200000):
    L=20
    x,y,theta = np.loadtxt('data/'+str(i)+'.txt',delimiter='\t',unpack=True)
    n = len(x)
    for i in range(n):
        for j in range(n):
            if(i!=j):
                r.append(np.sqrt((x[i]-x[j])**2+(y[i]-y[j])**2))
```

```python
counts, bin_edges = np.histogram(r, bins=150)
rdf = [counts[i]/bin_edges[i] for i in range(len(counts))]
```

```python
plt.plot(bin_edges[0:len(counts)],rdf)
```
