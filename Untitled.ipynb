```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('default')
#Extract Data
x1,y1 = np.loadtxt('vicsek40.txt',delimiter='\t',unpack=True)

x2,y2 = np.loadtxt('vicsek100.txt',delimiter='\t',unpack=True)

x3,y3 = np.loadtxt('vicsek400.txt',delimiter='\t',unpack=True)
x = np.hstack((x1,x2,x3))
y=np.hstack((y1,y2,y3))
plt.plot(x1,y1,'r')
plt.plot(x2,y2,'g')
plt.plot(x3,y3,'b')
```

```python
f = np.poly1d(np.polyfit(x2, y2, 4))
xnew = np.linspace(0, 4.75, num=41, endpoint=True)
plt.plot(x2, y2, 'o', xnew, f(xnew), 'b-')

f = np.poly1d(np.polyfit(x1, y1, 4))
xnew = np.linspace(0, 4.75, num=41, endpoint=True)
plt.plot(x1, y1, 'go', xnew, f(xnew), 'g-')

f = np.poly1d(np.polyfit(x3, y3, 4))
xnew = np.linspace(0, 4.75, num=41, endpoint=True)
plt.plot(x3, y3,'ro', xnew, f(xnew), 'r-')
```

```python
from scipy.interpolate import CubicSpline

x5,y5 = np.loadtxt('vicsek40s.txt',delimiter='\t',unpack=True)

f1 = CubicSpline(x5,y5)
#f = np.poly1d(np.polyfit(x5, y5, 4))
xnew = np.linspace(0, 7.75, num=100, endpoint=True)
plt.plot(x5, y5, 'bo', xnew, f1(xnew), 'b-')
```
