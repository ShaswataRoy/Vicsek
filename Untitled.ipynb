```python
x7,y7 = np.loadtxt('std/vicsek10000s.txt',delimiter='\t',unpack=True)
x8=0
y8=0
for i in range(10):
    s = 'std/vicsek1000'+str(i)+'s.txt'
    x7,y7 = np.loadtxt(s,delimiter='\t',unpack=True)
    x8+=x7
    y8+=y7
    
x8 = x8/10
y8 = y8/10
f5 = CubicSpline(x8,y8)
#f = np.poly1d(np.polyfit(x5, y5, 4))
xnew = np.linspace(0, 7.75, num=100, endpoint=True)
plt.plot(x8, y8, 'bo', xnew, f5(xnew), 'b-')
```

```python
xnew = np.linspace(0, 6.28, num=100, endpoint=True)

plt.plot(x5[0:25],10*y5[0:25],'o',xnew, 10*f1(xnew),label='N=40')
plt.plot(x7[0:25],15*y7[0:25],'o',xnew, 15*f2(xnew),label='N=60')
plt.plot(x8[0:25],20*y8[0:25],'o',xnew, 20*f3(xnew),label='N=80')
plt.plot(x6[0:25],25*y6[0:25],'o',xnew, 25*f4(xnew),label='N=100')
plt.legend()
plt.xlabel('Noise $\eta$')
plt.ylabel('Susceptibility $\chi$')
#plt.plot(xnew, 100*f5(xnew))
#plt.plot(x8, 250*y8, 'go', xnew, 250*f3(xnew), 'g-')
```
